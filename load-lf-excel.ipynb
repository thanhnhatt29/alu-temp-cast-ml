```python
import pandas as pd
import os

pd.set_option('display.max_columns', None)

# Define file paths
file_paths = [
    r"e:\OneDrive - hoaphat.com.vn\Code\alu-temp-cast-ml\01-data\LF\25.12.LF2.xlsx",
    r"e:\OneDrive - hoaphat.com.vn\Code\alu-temp-cast-ml\01-data\LF\25.12.LF5.xlsx"
]

# Define column names as requested
cols = ['stt', 'ngay', 'Ca', 'me_tinh_luyen_so', 'mac_thep_yeu_cau', 'thoi_gian_vao_tinh_luyen', 'bat_dau', 'ket_thuc', 'thoi_gian_len_duc', 'thung_lf', 'lan_luyen_thu', 'nhiet_do_vao_tl', 'C_truoc', 'Si_truoc', 'Mn_truoc', 'S_truoc', 'P_truoc', 'khoi_luong_thung_thep', 'FeSi', 'FeMn', 'SiMn', 'than', 'FeCr', 'FeV', 'Niken', 'FeP', 'Cu', 'khac', 'huynh_thach', 'nhom_thoi', 'voi_song', 'dolomite', 'quaczit', 'day_feca', 'day_casi', 'day_ca_dac', 'xi_bao_on', 'thoi_gian_danh_dien', 'tieu_thu_dien', 'C_sau', 'Si_sau', 'Mn_sau', 'S_sau', 'P_sau', 'Al', 'Canxi', 'nhiet_do_lan_1', 'nhiet_do_ra_thep', 'nhiet_do_duc_yeu_cau', 'nhiet_do_do_tren_duc', 'thoi_gian_dinh_tre', 'ly_do_dinh_tre', 'ghi_chu_1', 'thoi_gian_bat_dau_thoi_mem', 'thoi_gian_ket_thu_thoi_mem', 'tong_thoi_gian_thoi_mem', 'tinh_trang_xi_lo_thoi_qua_tinh_luyen', 'tinh_trang_xi', 'ghi_chu']
```

```python
# List to hold dataframes
dfs = []

for path in file_paths:
    if os.path.exists(path):
        print(f"Reading {path}...")
        try:
            # Read excel using calamine for speed, skiprows=4 to skip header rows
            df = pd.read_excel(path, skiprows=4, header=None, names=cols, engine='calamine')
            df = df.dropna(how='all')
            dfs.append(df)
            print(f"  -> Loaded {len(df)} rows.")
        except Exception as e:
            print(f"  -> Error reading file: {e}") 
    else:
        print(f"File not found: {path}")
```

```python
final_df = pd.concat(dfs, ignore_index=True)
df = final_df.copy()
```

```python
df = df.dropna(subset=['me_tinh_luyen_so'])
df = df[df['mac_thep_yeu_cau'].str.contains('SAE1006', na=False)]
df = df.drop(columns=['stt'])
```
