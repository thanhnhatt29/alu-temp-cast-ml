```python
import pandas as pd

# 2. Tính toán lượng dữ liệu thiếu
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

# 3. Tạo bảng tổng hợp để dễ nhìn
missing_data = pd.concat([missing_count, missing_percent], axis=1, keys=['Total Missing', 'Percent (%)'])

# 4. Chỉ hiển thị những cột có dữ liệu thiếu (lớn hơn 0)
print("Các cột đang bị thiếu dữ liệu:")
print(missing_data[missing_data['Total Missing'] > 0].sort_values(by='Total Missing', ascending=False))
```

```python
# 3. Tính giá trị phổ biến nhất (Top Value)
# Dùng vòng lặp để xử lý an toàn cho từng cột (tránh lỗi nếu cột rỗng)
top_values = {}
for col in df.columns:
    try:
        # Lấy giá trị xuất hiện nhiều nhất (bỏ qua NaN)
        top_val = df[col].value_counts().idxmax()
    except ValueError:
        # Trường hợp cột toàn NaN
        top_val = "N/A (Empty)"
    top_values[col] = top_val

# 4. Tạo bảng thống kê
unique_stats = pd.DataFrame({
    'Unique Count': df.nunique(),
    'Percent Unique (%)': (df.nunique() / len(df) * 100).round(4),
    'Data Type': df.dtypes,
    'Top Value': pd.Series(top_values) # Cột mới thêm vào
})

# 5. Sắp xếp: Cột ít giá trị nhất lên đầu (Dễ nhìn các cột cần xóa)
unique_stats = unique_stats.sort_values(by='Unique Count', ascending=True)

# 6. Hiển thị kết quả
print(f"Tổng số dòng dữ liệu: {len(df)}")
print("-" * 100)
# Dùng to_markdown() để canh lề bảng đẹp hơn khi in ra
unique_stats
```
