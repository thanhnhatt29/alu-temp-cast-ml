# Khám phá dữ liệu cơ bản

```python
# Xem các cột và thống kê mô tả
print("Tên các cột:")
print(df.columns.tolist())
print("\nThống kê mô tả cho các cột số:")
df.describe()
```

```python
# Kiểm tra giá trị null
print("Tỷ lệ giá trị null trong từng cột:")
null_percentage = (df.isnull().sum() / len(df)) * 100
null_info = null_percentage[null_percentage > 0].sort_values(ascending=False)
print(null_info.head(10))
```

```python
# Phân tích các cột quan trọng
print("=== TỔNG QUAN DỮ LIỆU SẢN XUẤT THÉP ===")
print(f"Tổng số mẫu: {len(df):,}")
print(f"Tổng số cột: {len(df.columns)}")
print(f"Khoảng thời gian: {df['ProductionDate'].min()} đến {df['ProductionDate'].max()}")
print()

# Phân tích các cột phân loại quan trọng
print("=== PHÂN LOẠI SẢN PHẨM ===")
print("Các loại sản phẩm (ClassifyName):")
classify_counts = df['ClassifyName'].value_counts(dropna=False)
print(classify_counts)
print()

print("Các cấp thép (ProductGrade):")
print(df['ProductGrade'].value_counts().head())
print()

print("Các ca làm việc (ShiftName):")
print(df['ShiftName'].value_counts())
print()

print("Các lò sản xuất (OvenCode):")
print(df['OvenCode'].value_counts())
```

```python
# Phân tích thành phần hóa học chính
print("=== THÀNH PHẦN HÓA HỌC CHÍNH ===")
chemical_cols = ['C', 'Si', 'Mn', 'S', 'P', 'Cu', 'Ni', 'Cr', 'Mo', 'V', 'Ti', 'Al']
chemical_summary = df[chemical_cols].describe().round(4)
print(chemical_summary[['C', 'Si', 'Mn', 'S', 'P']])
print()

# CEV analysis
print("=== CARBON EQUIVALENT VALUE (CEV) ===")
print(f"CEV trung bình: {df['CEV'].mean():.4f}")
print(f"CEV min: {df['CEV'].min():.4f}")
print(f"CEV max: {df['CEV'].max():.4f}")
print(f"Số mẫu có CEV: {df['CEV'].notna().sum():,}")
print()

# Phân tích mức độ thiếu dữ liệu
print("=== ĐÁNH GIÁ CHẤT LƯỢNG DỮ LIỆU ===")
missing_data = (df.isnull().sum() / len(df) * 100).round(2)
print("Các cột có dữ liệu đầy đủ (< 5% missing):")
complete_cols = missing_data[missing_data < 5].index.tolist()
print(f"Số lượng: {len(complete_cols)} cột")
print(complete_cols[:10])  # Show first 10
print()

print("Các cột thiếu nhiều dữ liệu (> 50% missing):")
high_missing = missing_data[missing_data > 50].sort_values(ascending=False)
print(f"Số lượng: {len(high_missing)} cột")
print(high_missing.head())
```

# Những gì bạn có thể làm với dữ liệu này

Dựa trên phân tích sơ bộ, đây là dữ liệu sản xuất thép với nhiều thông tin về:
- **Thông tin sản xuất**: ca làm việc, ngày sản xuất, cấp thép, kích thước, lò sản xuất
- **Thành phần hóa học**: Carbon (C), Silicon (Si), Manganese (Mn), và 40+ nguyên tố khác
- **Phân loại chất lượng**: ClassifyID, ClassifyName

## Các phân tích và mô hình có thể thực hiện:

### 1. **Kiểm soát chất lượng và phân loại**
- **Dự đoán chất lượng sản phẩm**: Sử dụng các đặc trưng hóa học để dự đoán ClassifyName (Loại 1, Loại 2, v.v.)
- **Phát hiện bất thường**: Tìm những mẫu có thành phần hóa học bất thường hoặc nằm ngoài tiêu chuẩn

### 2. **Tối ưu hóa quy trình sản xuất**
- **Phân tích hiệu suất theo ca**: So sánh chất lượng sản phẩm giữa các ca làm việc
- **Tối ưu hóa lò sản xuất**: Phân tích hiệu suất theo từng lò (OvenID, OvenCode)
- **Dự đoán thời gian sản xuất**: Dựa vào thành phần để ước tính thời gian cần thiết

### 3. **Phân tích thành phần hóa học**
- **Tìm hiểu mối quan hệ giữa các nguyên tố**: Correlation matrix, PCA để giảm chiều dữ liệu
- **Tối ưu công thức thép**: Tìm tỷ lệ nguyên tố tốt nhất cho từng loại sản phẩm
- **Phân tích CEV (Carbon Equivalent Value)**: Dự đoán tính hàn của thép

### 4. **Machine Learning Models**
- **Classification**: Dự đoán loại sản phẩm (ClassifyName) từ thành phần hóa học
- **Regression**: Dự đoán các tính chất cơ học từ thành phần
- **Clustering**: Nhóm sản phẩm theo đặc tính tương tự
- **Time Series**: Phân tích xu hướng chất lượng theo thời gian

### 5. **Báo cáo và Dashboard**
- **Dashboard theo dõi chất lượng**: Real-time monitoring
- **Báo cáo hiệu suất sản xuất**: Theo ca, theo ngày, theo tháng
- **Cảnh báo sớm**: Khi thành phần vượt ngưỡng cho phép
